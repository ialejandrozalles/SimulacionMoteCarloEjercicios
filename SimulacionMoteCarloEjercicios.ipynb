Sea (X,Y)(X,Y) es un vector aleatorio con distribución uniforme en el cuadrado [−1,1]×[−1,1][−1,1]×[−1,1] de área 4.

a) Aproximar mediante simulación P(X+Y≤0)P(X+Y≤0) y compararla con la probabilidad teórica (obtenida aplicando la regla de Laplace área favorable/área posible..

b) Aproximar el valor de π mediante simulación a partir de P(X2+Y2≤1).

c) Generamos el correspondiente gráfico (ver Figura ) (los puntos con color negro tienen distribución uniforme en el círculo unidad; esto está relacionado con el método de aceptación-rechazo).

```r
numeroPuntos <- 100000
```

```r
generarPuntosAleatorios <- function(n) {
  x <- runif(n, min = -1, max = 1)
  y <- runif(n, min = -1, max = 1)
  return(list(x = x, y = y))
}
```

```r
# a) Aproximar P(X+Y≤0) mediante simulación y comparar con la teórica
calcularProbabilidadXmasYMenorIgualCero <- function(puntos) {
  x <- puntos$x
  y <- puntos$y
  condicion <- x + y <= 0
  probSimulada <- mean(condicion)

  probTeorica <- 0.5

  return(list(simulada = probSimulada, teorica = probTeorica))
}
```

```r
# b) Aproximar el valor de π mediante P(X²+Y²≤1)
aproximarPi <- function(puntos) {
  x <- puntos$x
  y <- puntos$y
  condicion <- x^2 + y^2 <= 1
  probCirculo <- mean(condicion)
  piAproximado <- 4 * probCirculo

  return(piAproximado)
}
```

```r
# c) Generar gráfico similar al proporcionado
generarGrafico <- function(puntos) {
  x <- puntos$x
  y <- puntos$y
  dentroCirculo <- x^2 + y^2 <= 1

  plot(x, y, pch = 20, cex = 0.5, col = ifelse(dentroCirculo, "black", "red"),
       xlim = c(-1.1, 1.1), ylim = c(-1.1, 1.1),
       xlab = "X", ylab = "Y", main = "Simulación de puntos en cuadrado [-1,1]×[-1,1]")

  theta <- seq(0, 2 * pi, length.out = 200)
  lines(cos(theta), sin(theta))
}
```

```r
# Ejecutar simulaciones
set.seed(123)
puntos <- generarPuntosAleatorios(numeroPuntos)
```

```r
# Parte a
resultadoA <- calcularProbabilidadXmasYMenorIgualCero(puntos)
cat("a) P(X+Y≤0):\n")
cat("   Probabilidad simulada:", resultadoA$simulada, "\n")
cat("   Probabilidad teórica:", resultadoA$teorica, "\n")
cat("   Diferencia:", abs(resultadoA$simulada - resultadoA$teorica), "\n\n")
```

##### La simulación es precisa la diferencia con el valor teórico es mínima y valida el buen uso de números pseudoaleatorios

```r
# Parte b
piAproximado <- aproximarPi(puntos)
cat("b) Aproximación de π mediante simulación:\n")
cat("   Valor aproximado:", piAproximado, "\n")
cat("   Valor real de π:", pi, "\n")
cat("   Error relativo:", abs(piAproximado - pi)/pi * 100, "%\n\n")
```

##### La simulación aproxima bien el valor de π el error relativo es bajo y demuestra la eficacia del método usado

```r
# Parte c
cat("c) Grafico:\n\n")
generarGrafico(puntos)
```

##### El gráfico muestra cómo se distribuyen los puntos aleatorios dentro del cuadrado y el círculo negro dentro visualiza cómo se aproxima π comparando áreas mediante simulación
